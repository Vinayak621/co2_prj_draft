```python
d['classify'] = np.where((d['cereal_yield'] > 2584.8) &  ((d['fdi_perc_gdp']>2)| (d['en_per_gdp'] > 197.62) |  (d['gdp'] > 98829417855))
                             & ((d['co2_ttl'] > 77283.8585)| (d['co2_per_cap']>3)|(d['co2_per_gdp']>368)) &  ((d['prot_area_perc'] > 10 )| (d['pop_urb_aggl_perc'] > 19)|(d['pop']>26892071)|(d['urb_pop']>14232388)) & (d['gni_per_cap'] > 2151), 'red', 'green')
```

```python
X = d.iloc[:, :-1].values
y = d.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV, RandomizedSearchCV
from sklearn.svm import SVC
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Support Vector Machine and it's differnt kernels

```python
from sklearn.svm import SVC
classifier2 = SVC(kernel = 'linear', random_state = 0)
classifier2.fit(X_train, y_train)
```

```python
y_pred = classifier2.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
poly_svm = SVC(kernel='poly', degree=3, coef0=1)
poly_svm.fit(X_train, y_train)
y_pred1 = poly_svm.predict(X_test)
poly_accuracy = accuracy_score(y_test, y_pred1)
print("Accuracy with polynomial kernel:", poly_accuracy)
cm1 = confusion_matrix(y_test, y_pred1)
print(cm1)
accuracy_score(y_test, y_pred1)
```

```python
rbf_svm = SVC(kernel='rbf', gamma='scale')
rbf_svm.fit(X_train, y_train)
y_pred2 = rbf_svm.predict(X_test)
rbf_accuracy = accuracy_score(y_test, y_pred2)
print("Accuracy with RBF kernel:", rbf_accuracy)
```

```python
sigmoid_svm = SVC(kernel='sigmoid', coef0=1)
sigmoid_svm.fit(X_train, y_train)
y_pred3 = sigmoid_svm.predict(X_test)
sigmoid_accuracy = accuracy_score(y_test, y_pred3)
print("Accuracy with sigmoid kernel:", sigmoid_accuracy)
```
